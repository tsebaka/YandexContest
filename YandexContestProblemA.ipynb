# **Libraries**

```python
import numpy as np
import pandas as pd
from math import *
import matplotlib.pyplot as plt
```

# **Functions**

```python
def generate1():  
    a = np.random.uniform(0, 1) 
    b = np.random.uniform(0, 1) 
    return (a * cos(2 * pi * b), a * sin(2 * pi * b))
def generate2():  
    while True:  
        x = np.random.uniform(-1, 1)  
        y = np.random.uniform(-1, 1)  
        if x ** 2 + y ** 2 > 1:  
            continue  
        return (x, y)
```

# **Data**

```python
count_1_gen = 0
count_2_gen = 0

for i in range (1):
  generator_1_x1 = []
  generator_1_y1 = []
  generator_2_x2 = []
  generator_2_y2 = []
  for j in range(1000):
    x1, y1 = generate1()
    x2, y2 = generate2()
    generator_1_x1.append(x1)
    generator_1_y1.append(y1)
    generator_2_x2.append(x2)
    generator_2_y2.append(y2)
    if x1 <= 0.25 and x1 >= -0.25 and y1 >= -0.25 and y1 <= 0.25:
      count_1_gen += 1
    if x2 <= 0.25 and x2 >= -0.25 and y2 >= -0.25 and y2 <= 0.25:
      count_2_gen += 1
```

# **Graph analysis**

```python
print("First generator:", count_1_gen / 100)
plt.plot(generator_1_x1, generator_1_y1, 'ro', color = 'green')
plt.show()

print("Second generator:", count_2_gen / 100)
plt.plot(generator_2_x2, generator_2_y2, 'ro', color = 'blue')
plt.show()
```
